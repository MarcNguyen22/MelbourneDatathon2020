```python
import pandas as pd
import datetime as dt
```

```python
data_2014_15=pd.read_csv("public_2014_2015.csv")
data_2016_17=pd.read_csv("public_2016_2017.csv")
data_2018=pd.read_csv("public_2018.csv")
data_2019=pd.read_csv("public_2019.csv")
data_2020=pd.read_csv("public_2020.csv")
```

```python
print(data_2014_15.columns)
print(data_2016_17.columns)
print(data_2018.columns)
print(data_2019.columns)
print(data_2020.columns)
```

```python
data_2019=data_2019.drop(["Raw Date"],axis=1)
```

```python
data_2019=data_2019.rename(columns={"Jurisdiction":"Applicable To"})
data_2020=data_2020.rename(columns={"Jurisdiction":"Applicable To"})
```

```python
public_holidays=pd.concat([data_2014_15,data_2016_17,data_2018,data_2019,data_2020])
```

```python
public_holidays
```

```python
public_holidays=public_holidays.drop(["Information","More Information"],axis=1)
```

```python
vic_ph=public_holidays[public_holidays["Applicable To"].str.contains("|".join(["v","V","NAT"]))]
```

```python
vic_ph=vic_ph.drop("Applicable To",axis=1)
```

```python
vic_ph
```

```python
vic_ph["Year"]=vic_ph.apply(lambda x:int(str(x.Date)[:4]),axis=1)
vic_ph["Month"]=vic_ph.apply(lambda x:int(str(x.Date)[4:6]),axis=1)
vic_ph["Day"]=vic_ph.apply(lambda x:int(str(x.Date)[6:]),axis=1)
```

```python
vic_ph
```

```python
school_hols=pd.read_csv("school_holidays.csv")
```

```python
school_hols
```

```python
base = dt.datetime(2014,1,1)
date_list = [base + dt.timedelta(days=x) for x in range(2557)]
```

```python
date_list[-1]
```

```python
dt.datetime.strptime(school_hols.iloc[1].start_date,'%d/%m/%Y')
```

```python
date_list[0].weekday()
```

```python
dt.datetime.strptime(school_hols.iloc[1].start_date,'%d/%m/%Y')<=date_list[27]<=dt.datetime.strptime(school_hols.iloc[1].end_date,'%d/%m/%Y')
```

```python
j=0
Weekdays=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
hol_array=[]
day_array=[]
year_array=[]
month_array=[]
date_array=[]
weekday_int_array=[]
for single_date in date_list:
    year_array.append(single_date.year)
    month_array.append(single_date.month)
    date_array.append(single_date.day)
    day_array.append(Weekdays[single_date.weekday()])
    weekday_int_array.append(single_date.weekday())
    df=vic_ph[(vic_ph.Month==single_date.month)&(vic_ph.Day==single_date.day)&(vic_ph.Year==single_date.year)]
    if len(df)!=0:
        hol_array.append("PublicHoliday")
    else:
        i=0
        datebool=False
        while i<len(school_hols):
            if dt.datetime.strptime(school_hols.iloc[i].start_date,'%d/%m/%Y')<=single_date<=dt.datetime.strptime(school_hols.iloc[i].end_date,'%d/%m/%Y'):
                hol_array.append("SchoolHoliday")
                datebool=True
                break
            i=i+1
        if not datebool:
            hol_array.append("Normal")
    j=j+1
```

```python
Date_df=pd.DataFrame()
Date_df["Year"]=year_array
Date_df["Month"]=month_array
Date_df["Day"]=date_array
Date_df["Weekday"]=day_array
Date_df["Weekday_int"]=weekday_int_array
Date_df["Seasonality"]=hol_array
Date_df
```

```python
Date_df.to_csv("date_dim.csv",index=False)
```
