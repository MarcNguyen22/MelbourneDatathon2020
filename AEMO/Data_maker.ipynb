```python
#set up the dataframe
coltitles=['REGION', 'SETTLEMENTDATE', 'TOTALDEMAND', 'RRP', 'PERIODTYPE']
df=pd.DataFrame(columns=coltitles)

#set up the values to iterate over
States=["NSW","VIC","TAS","QLD","SA"]
Years=range(1998,2021)
Months=["01","02","03","04","05","06","07","08","09","10","11","12"]
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36'}
```

```python
#Obtain the data
for year in Years:
    for month in Months:
        for state in States:
            try:
                url="https://aemo.com.au/aemo/data/nem/priceanddemand/PRICE_AND_DEMAND_{}{}_{}1.csv".format(year,month,state)
                data=requests.get(url, headers= headers).text
                temp_df=pd.read_csv(StringIO(data))
                df=pd.concat([df,temp_df])
                print("{}, {}, {}: Complete".format(year,month,state))
            except:
                continue
```
