```python
import pandas as pd
```

```python
mintemp=pd.read_csv("Mel_Min_temp.csv")
maxtemp=pd.read_csv("Mel_Max.csv")
solar=pd.read_csv("Mel_Solar.csv")
```

```python
mintemp
```

```python
maxtemp
```

```python
tempcombined=mintemp.merge(maxtemp,how="inner",on=["Year","Month","Day"])
```

```python
tempcombined
```

```python
data=tempcombined.merge(solar,how="inner",on=["Year","Month","Day"])
```

```python
data.isna().sum()
```

```python
data=data.dropna()
```

```python
data=data[["Year","Month","Day","Minimum temperature (Degree C)","Maximum temperature (Degree C)","Daily global solar exposure (MJ/m*m)"]]
```

```python
data
```

```python
data.to_csv("Mel_weather.csv",index=False)
```
